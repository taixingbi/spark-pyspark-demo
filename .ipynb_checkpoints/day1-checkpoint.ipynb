```python
import sparknlp
sparknlp.start()

from sparknlp.pretrained import PretrainedPipeline
```

```python
def pre_train_pipeline(pipeline, document):
    explain_document_pipeline = PretrainedPipeline(pipeline)
    annotations = explain_document_pipeline.annotate(document)
    print(annotations)
```

```python
document= "We are very happy about SparkNLP. this is hunter test tomorrow"
```

```python
pre_train_pipeline("explain_document_ml", document)
```

```python
pre_train_pipeline("explain_document_dl", document)
```
